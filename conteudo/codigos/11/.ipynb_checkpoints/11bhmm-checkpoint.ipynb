## Algoritmo Forward

```python
def forward(y, states, a_0, a, e):
    L = len(y)
 
    fwd = []
    f_prev = {}

    for i, y_i in enumerate(y):
        f_curr = {}
        for st in states:
            if i == 0:
                # caso-base (prob a priori)
                prev_f_sum = a_0[st]
            else:
                prev_f_sum = sum(f_prev[k]*a[k][st] for k in states)
 
            f_curr[st] = e[st][y_i] * prev_f_sum
 
        fwd.append(f_curr)
        f_prev = f_curr
 
    posterior = []
    for i in range(L):
        norm_sum = 0
        for st in states:
            norm_sum = norm_sum + fwd[i][st]
        norm_factor = 1.0/norm_sum
        posterior.append({st: fwd[i][st]*norm_factor for st in states})   

    return posterior
```

### Aplicando o algoritmo Forward para resolver o problema do guarda-chuva

```python
states = ('Rain', 'Sun')
#end_state = 'E'
 
# evidências
observations = ('umbrella_false', 'umbrella_true', 'umbrella_false')

# Probabilidades a priori
start_probability = {'Rain': 0.5, 'Sun': 0.5}

# Modelo de transição
transition_probability = {
   'Rain' : {'Rain': 0.7, 'Sun': 0.3},
   'Sun' : {'Rain': 0.3, 'Sun': 0.7},
   }

# Modelo sensor 
emission_probability = {
   'Rain' : {'umbrella_true': 0.9, 'umbrella_false': 0.1},
   'Sun' : {'umbrella_true': 0.2, 'umbrella_false': 0.8},
   }
```

```python
def example():
    return forward(observations,
                   states,
                   start_probability,
                   transition_probability,
                   emission_probability)
 
prob = example()
print (prob)
```

### Aplicando o algoritmo Forward para resolver o problema da vila

(Exemplo wikipedia) Considere um vilarejo onde os moradores estão saudáveis ou possuem um tipo de febre e um único médico disponível. O médico diagnostic a febre perguntando aos pacientes como se sentem. Os pacientes respondem apenas se se sentem normal, tontos (dizzy) ou com frio (cold). O médico acredita que o estado de saúde dos pacientes opera como uma cadeia de Markov discreta com dois possiveis valores para o estado do paciente: "Healthy" e "Fever". Mas o médico não pode observá-los diretamente; estão escondidos. A cada dia, existe alguma chance do paciente apontar para o médico um dos três sintomas. Este conjunto de observações possíveis (normal, cold, dizzy) juntamente com este estado oculto (healthy, fever) formam um hidden Markov model (HMM).

```python
from IPython.display import Image

Image(filename='./vila.png', width=400) 
```

Uma moradora visita por três dias consecultivos o médico e apresenta a seguinte sequência de sintomas: normal, cold e dizzy.<br>

O médico então se pergunta: qual a chance dela estar doente?

```python
states = ('Healthy', 'Fever')
#end_state = 'E'
 
# evidências
observations = ('normal', 'cold', 'dizzy')

# Probabilidades a priori
start_probability = {'Healthy': 0.6, 'Fever': 0.4}

# Modelo de transição
transition_probability = {
   'Healthy' : {'Healthy': 0.7, 'Fever': 0.3},
   'Fever' : {'Healthy': 0.4, 'Fever': 0.6},
   }

# Modelo sensor 
emission_probability = {
   'Healthy' : {'normal': 0.5, 'cold': 0.4, 'dizzy': 0.1},
   'Fever' : {'normal': 0.1, 'cold': 0.3, 'dizzy': 0.6},
   }
```

```python
def example():
    return forward(observations,
                   states,
                   start_probability,
                   transition_probability,
                   emission_probability)
 
prob = example()
print (prob)
```

# Decodificação (decoding)
Serve para resolver o problema de encontrar a sequência de estados ocultos que melhor explica a sequencia de observacões.

## Algoritmo de Viterbi

```python
def print_dptable(V):
    s = "    " + " ".join(("%7d" % i) for i in range(len(V))) + "\n"
    for y in V[0]:
        s += "%.5s: " % y
        s += " ".join("%.7s" % ("%f" % v[y]) for v in V)
        s += "\n"
    print(s)

def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]
    path = {}
 
    # Inicializa casos-base (t == 0)
    for y in states:
        V[0][y] = start_p[y] * emit_p[y][obs[0]]
        path[y] = [y]
 
    # Executa Viterbi para t > 0
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}
 
        for y in states:
            (prob, state) = max((V[t-1][y0] * trans_p[y0][y] * emit_p[y][obs[t]], y0) for y0 in states)
            V[t][y] = prob
            newpath[y] = path[state] + [y]
 
        # não se faz necessário lembrar de caminhos antigos
        path = newpath
     
    #print_dptable(V)
    (prob, state) = max((V[t][y], y) for y in states)
    
    return (V, path[state])
```

### Aplicando o algoritmo de Viterbi para resolver o problema da vila

Uma moradora visita por três dias consecultivos o médico e apresenta a seguinte sequência de sintomas: normal, cold e dizzy.<br>

O médico então se pergunta: qual a sequência de estados de saúde dessa paciente que melhor explicaria essa sequência de sintomas apresentados?

```python
states = ('Healthy', 'Fever')
#end_state = 'E'
 
# evidências
observations = ('normal', 'cold','dizzy')

# Probabilidades a priori
start_probability = {'Healthy': 0.6, 'Fever': 0.4}

# Modelo de transição
transition_probability = {
   'Healthy' : {'Healthy': 0.7, 'Fever': 0.3},
   'Fever' : {'Healthy': 0.4, 'Fever': 0.6},
   }

# Modelo sensor 
emission_probability = {
   'Healthy' : {'normal': 0.5, 'cold': 0.4, 'dizzy': 0.1},
   'Fever' : {'normal': 0.1, 'cold': 0.3, 'dizzy': 0.6},
   }
```

```python
# input the generated markov model
def example():
    return viterbi(observations,
                   states,
                   start_probability,
                   transition_probability,
                   emission_probability)

(v, path_hidden) = example()

print (v)
print ('Sequência mais provável:',path_hidden)
```
